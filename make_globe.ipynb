```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python

ax = plt.axes(projection=ccrs.NearsidePerspective(central_latitude=80.0,central_longitude=-90))
ax.stock_img()
plt.show()
```

```python
ax = plt.axes(projection=ccrs.NearsidePerspective(central_latitude=80.0,central_longitude=-87.6))
ax.stock_img()



chi_lon, chi_lat = -87.6, 41.8
oslo_lon, oslo_lat = 10.8, 59.9
tokyo_lon, tokyo_lat = 139.7,35.7
seoul_lon, seoul_lat = 127.0,37.6
beijing_lon, beijing_lat = 116.4, 39.9 


plt.plot([chi_lon, beijing_lon], [chi_lat, beijing_lat],
         color='red', linewidth=2, marker='o',
         transform=ccrs.Geodetic(),
         )


#plt.plot([chi_lon, tokyo_lon], [chi_lat, tokyo_lat],
#         color='blue', linewidth=2, marker='o',
#         transform=ccrs.Geodetic(),
#         )

plt.plot([chi_lon, seoul_lon], [chi_lat, seoul_lat],
         color='black', linewidth=2, marker='o',
         transform=ccrs.Geodetic(),
         )


plt.plot([chi_lon, oslo_lon], [chi_lat, oslo_lat],
         color='blue', linewidth=2, marker='o',
         transform=ccrs.Geodetic(),
         )


plt.plot(np.linspace(0,359,360), np.ones(360)*66.5,
         color='black', linewidth=2, ls='--',
         transform=ccrs.Geodetic(),
         )



plt.text(0, 90, '+',
         horizontalalignment='right',
         transform=ccrs.Geodetic())


plt.text(chi_lon - 3, chi_lat - 12, 'Chicago',
         horizontalalignment='right',
         transform=ccrs.Geodetic())

#plt.text(tokyo_lon, tokyo_lat - 20, 'Tokyo',
#         horizontalalignment='right',
#         transform=ccrs.Geodetic())

plt.text(seoul_lon, seoul_lat - 20, 'Seoul',
         horizontalalignment='right',
         transform=ccrs.Geodetic())

plt.text(beijing_lon-10, beijing_lat - 15, 'Beijing',
         horizontalalignment='right',
         transform=ccrs.Geodetic())

plt.text(oslo_lon - 3, oslo_lat - 4, 'Oslo',
         horizontalalignment='left',
         transform=ccrs.Geodetic())



plt.savefig('globe.png')
```

```python
import numpy as np
plt.figure()
time = np.linspace(2018,2020,25)
fare1 = 1400+300*(time-2018.6)*np.sin(time*2.9+3.7)
fare2 = 1200+300*(time-2018.6)*np.sin(time*4.9+3.7)
fare3 = 1000+400*np.sin(time*6.0+11)*(time-2017.0)**0.5


plt.xticks([2018,2018.5,2019,2019.5,2020],['Jan 2018','Jun 2018','Jan 2019','Jun 2019','Jan 2020'],size=14)
p1,=plt.plot(time,fare1,'-k',lw=4,alpha=0.6)
p2,=plt.plot(time,fare2,'-r',lw=4,alpha=0.6)
p3,=plt.plot(time,fare3,'-b',lw=4,alpha=0.6)
plt.ylim([0,2500])

plt.legend([p1,p2,p3],['Seoul','Beijing','Oslo'])


plt.ylabel('estimated average airfare (USD)',size=14)
plt.savefig('ex_fare.pdf')
```

```python
plt.figure()
time = np.linspace(2018,2020,25)
prob1 = (time-2017)**-0.5 * (0.4+0.3*np.sin(time*6.2))
prob2 = (time-2017)**-0.5 * (0.35+0.245*np.sin(time*6.2))
prob3 = (time-2017)**-0.5 * (0.2+0.1*np.sin(time*6.2))


plt.xticks([2018,2018.5,2019,2019.5,2020],['Jan 2018','Jun 2018','Jan 2019','Jun 2019','Jan 2020'],size=14)
p1,=plt.plot(time,prob1,'-k',lw=4,alpha=0.6)
p2,=plt.plot(time,prob2,'-r',lw=4,alpha=0.6)
p3,=plt.plot(time,prob3,'-b',lw=4,alpha=0.6)
plt.legend([p1,p2,p3],['Seoul','Beijing','Oslo'])
plt.ylim([0,1])


plt.ylabel('aurora probability',size=14)
plt.savefig('ex_prob.pdf')
```
